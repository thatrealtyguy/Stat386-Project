```python
gun_data['Element'] = np.zeros(len(gun_data))
gun_data['Type'] = np.zeros(len(gun_data))
gun_data['Rarity'] = np.zeros(len(gun_data))
```

```python
def get_last_word(text):
  words = text.split()
  return words[-1]

def remove_last_word(text):
  words = text.split()
  return ' '.join(words[:-1])

def insert_spaces(text):
  result = ""
  for i in range(len(text)):
    char = text[i]
    if i < len(text) - 1 and char.islower() and text[i + 1].isupper():
      result += char + " "
    elif i < len(text) - 1 and char.isdigit() and text[i + 1].isupper():
      result += char + " "
    elif i < len(text) - 1 and char == 'I' and text[i + 1] == '.':
      result += char + " "
    elif i < len(text) - 1 and char.isupper() and text[i + 1].isupper():
      result += char + " "
    else:
      result += char
  return result

def get_gun_type(text):
  word_list = ['Sword', 'Sidearm', 'Glaive', 'Shotgun', 'Hand Cannon',
               'Scout Rifle', 'Auto Rifle', 'Pulse Rifle', 'Submachine Gun',
               'Combat Bow', 'Fusion Rifle', 'Sniper Rifle', 'Trace Rifle',
               'Grenade Launcher', 'Machine Gun', 'Rocket Launcher',
               'Linear Fusion Rifle']
  words = text.split()
  for phrase in reversed(word_list):
    if words[-len(phrase.split()) :] == phrase.split():
      return phrase
  return None

def remove_gun_type(text):
  word_list = ['Sword', 'Sidearm', 'Glaive', 'Shotgun', 'Hand Cannon',
               'Scout Rifle', 'Auto Rifle', 'Pulse Rifle', 'Submachine Gun',
               'Combat Bow', 'Fusion Rifle', 'Sniper Rifle', 'Trace Rifle',
               'Grenade Launcher', 'Machine Gun', 'Rocket Launcher',
               'Linear Fusion Rifle']
  words = text.split()
  for phrase in reversed(word_list):
    if words[-len(phrase.split()) :] == phrase.split():
      num_words_to_remove = len(phrase.split())
      return " ".join(words[:-num_words_to_remove])
  return text
```

```python
gun_data['Name'] = gun_data['Name'].apply(insert_spaces)
gun_data['Element'] = gun_data['Name'].apply(get_last_word)
gun_data['Name'] = gun_data['Name'].apply(remove_last_word)
```

```python
gun_data['Type'] = gun_data['Name'].apply(get_gun_type)
gun_data['Name'] = gun_data['Name'].apply(remove_gun_type)
```

```python
gun_data['Rarity'] = gun_data['Name'].apply(get_last_word)
gun_data['Name'] = gun_data['Name'].apply(remove_last_word)
```

```python
gun_data = gun_data.dropna(subset=['Type'])
```
